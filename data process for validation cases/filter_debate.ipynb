```python
data=df[["candidate","sentiment","sentiment_confidence","tweet_created","tweet_location"]]
```

```python
Trump=data.loc[df["candidate"]=="Donald Trump"]
```

```python
Cruz=data.loc[df["candidate"]=="Ted Cruz"]
```

```python
Bush=data.loc[df["candidate"]=="Jeb Bush"]
```

```python
def multiplier(tag,value):
    if (tag=="Positive"):
        return value
    elif tag=="Negative":
        return -value
    else:
        return 0
```

```python
T_scores=list(map(multiplier,Trump.loc[:,"sentiment"],Trump.loc[:,"sentiment_confidence"]))
C_scores=list(map(multiplier,Cruz.loc[:,"sentiment"],Cruz.loc[:,"sentiment_confidence"]))
B_scores=list(map(multiplier,Bush.loc[:,"sentiment"],Bush.loc[:,"sentiment_confidence"]))
```

```python
Trump["scores"]=T_scores
```

```python
Cruz["scores"]=C_scores
Bush["scores"]=B_scores
```
